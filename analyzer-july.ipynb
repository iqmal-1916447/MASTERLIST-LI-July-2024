function input month name, return month integer

```python
def get_month_int(month_name):
    # Convert the month name to the corresponding month integer
    month_int = datetime.strptime(month_name, '%B').month
    return month_int
```

Perform filter

```python
region_list = ['Petaling', 'Klang', 'Gombak', 'Kuala Lumpur', 'Hulu Langat', 'Hulu Selangor', 'Sepang', 'Kuala Langat', 'Kuala Selangor', 'Sabak Bernam']
```

```python
# contractor filtering
data_df_contractor = data_df_contractor[data_df_contractor['Date/Time Opened'].dt.month == get_month_int(file_month_name)]
data_df_contractor = data_df_contractor[data_df_contractor['Case Origin'] == 'Ais Leak']
data_df_contractor = data_df_contractor[data_df_contractor['Work Type: Work Type Name'].str.contains(r'(repair works?|dlp)', case=False, regex=True, na=False)]
data_df_contractor = data_df_contractor[~( (data_df_contractor['Status'] == 'New') | (data_df_contractor['Status'].isna()) )]
data_df_contractor = data_df_contractor[data_df_contractor['Region'].isin(region_list)]
data_df_contractor = data_df_contractor[data_df_contractor['Category'].isin(['Water Complaint', 'Maintenance Complaint'])]
```

```python
# inhouse filtering
data_df_inhouse = data_df_inhouse[data_df_inhouse['Date/Time Opened'].dt.month == get_month_int(file_month_name)]
data_df_inhouse = data_df_inhouse[data_df_inhouse['Case Origin'] == 'Ais Leak']
data_df_inhouse = data_df_inhouse[data_df_inhouse['Work Type: Work Type Name'].str.contains(r'site verifications?', case=False, regex=True, na=False)]
data_df_inhouse = data_df_inhouse[~( (data_df_inhouse['Status'] == 'New') | (data_df_inhouse['Status'].isna()) )]
data_df_inhouse = data_df_inhouse[data_df_inhouse['Region'].isin(region_list)]
data_df_inhouse = data_df_inhouse[~data_df_inhouse['Action By'].str.contains(r'contractors?', case=False, regex=True, na=False)]    # omit action by contractor
data_df_inhouse = data_df_inhouse[data_df_inhouse['Category'].isin(['Water Complaint', 'Maintenance Complaint'])]
```

```python
# all aisleak filtering
data_df_all = data_df_all[data_df_all['Date/Time Opened'].dt.month == get_month_int(file_month_name)]
data_df_all = data_df_all[data_df_all['Case Origin'] == 'Ais Leak']
data_df_all = data_df_all[data_df_all['Region'].isin(region_list)]
```
