```python
# Resolving Problem of intersection case number above
def resolving_intersect(df_inhouse, df_contractor, priority_func):

    # Add new column SELECT
    df_inhouse['SELECTED'] = True   # default value
    df_contractor['SELECTED'] = True    # default value

    # get common case number 
    case_number = pd.merge(left=df_inhouse[['Case Number']], right=df_contractor[['Case Number']], on=['Case Number'], how='inner')
    case_number = case_number['Case Number'].tolist()

    # set case number as index for ease of access
    df_inhouse.set_index('Case Number', inplace=True)
    df_contractor.set_index('Case Number', inplace=True)

    # select related row
    # df_inhouse_related = df_inhouse[df_inhouse['Case Number'].isin(case_number)]
    # df_contractor_related = df_contractor[df_contractor['Case Number'].isin(case_number)]

    df_inhouse['priority'] = df_inhouse['Status'].apply(priority_func)
    df_contractor['priority'] = df_contractor['Status'].apply(priority_func)

    
    # compare
    
        
    for each in case_number:

        # display(inhouse_row)
        # display(contractor_row)

        if df_inhouse.loc[each,'priority'] == df_contractor.loc[each,'priority']:
            continue

        if df_inhouse.loc[each,'priority'] > df_contractor.loc[each,'priority']:        # large priority value means less important
            df_inhouse.loc[each,'SELECTED'] = False
        elif df_inhouse.loc[each,'priority'] < df_contractor.loc[each,'priority']:
            df_contractor.loc[each, 'SELECTED'] = False
        
    df_inhouse.drop(columns=['priority'], inplace=True)
    df_contractor.drop(columns=['priority'], inplace=True)

    df_inhouse = df_inhouse[df_inhouse['SELECTED'] == True]
    df_contractor = df_contractor[df_contractor['SELECTED'] == True]

    df_inhouse.drop(columns=['SELECTED'], inplace=True)
    df_contractor.drop(columns=['SELECTED'], inplace=True)

    # reset index
    df_inhouse.reset_index(inplace=True)
    df_contractor.reset_index(inplace=True)

    return df_inhouse, df_contractor

```
